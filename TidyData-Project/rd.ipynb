```python
import pandas as pd 
fed_rd = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2019/2019-02-12/fed_r_d_spending.csv")
```

```python
print(fed_rd)
```

```python
pivot_table = fed_rd.pivot_table(values="rd_budget", index="department", columns="year", aggfunc="sum")

# Display the pivot table
print("\nTotal R&D Budget per Department (Rows) per Year (Columns):\n", pivot_table)
```
